```python
#https://stackoverflow.com/questions/41192424/python-how-to-correct-misspelled-names
import difflib
def is_similar(first, second, ratio):
    return difflib.SequenceMatcher(None, first, second).ratio() > ratio
```

```python
#just to know how much unique country are present
correct_country_unique = correct_lower['country'].unique()
```

```python
result_wrong=set()
result_right=set()
```

```python
for i,j in misspelt_lower_array:#i=city name and j = country name for missplet words
    for k,l in correct_lower_array:#k=city name and l = country name for correct words
        if(j==l):#first comparing the cities name to reducing the no of comparison
            if(is_similar(i,k, 0.8)):#in the case study it was written that 20 % characters has replaced , thats why ratio=0.8
                result_wrong.add(i)
                result_right.add(k)
                
    
    
    
```
